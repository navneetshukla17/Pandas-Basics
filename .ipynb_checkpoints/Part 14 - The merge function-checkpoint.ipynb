```python
import numpy as np
import pandas as pd
```

```python
delivery = pd.read_csv('deliveries.csv')
delivery
```

```python
match = pd.read_csv('matches - matches.csv')
match
```

```python
merged_df = delivery.merge(match, left_on='match_id', right_on='id')
merged_df
```

```python
print(delivery.shape)
print(match.shape)
```

```python
merged_df.shape
```

```python
merged_df.groupby(['season', 'batsman'])['batsman_runs'].sum()
```

```python
type(merged_df.groupby(['season', 'batsman'])['batsman_runs'].sum())
```

```python
# Converting Series into a DataFrame 
orange_cap_holders = merged_df.groupby(['season', 'batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates('season', keep='first').sort_values('season')[['season', 'batsman']]
```
