```python
# Creating a function using this data.
def get_city_match_count(city):
    '''
    This function displays the number of matches played in specific city.
    '''
    mask = df['city'] == city
    return df[mask].shape[0]


# calling the function:
get_city_match_count('Mumbai')
```

```python
# Two conditions at a time
mask1 = df['city'] == 'Mumbai'
mask2 = df['date'] > '2010-01-01'
```
