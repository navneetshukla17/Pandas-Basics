```python
# Single index pivot table
food.pivot_table(index='City', columns='Item', values='Spends', aggfunc='mean') 
```

```python
# multiple index pivot table
food.pivot_table(index=['City', 'Gender'], columns='Item', values='Spends', aggfunc='mean') 
```

```python
# adding one more dimension to coumns
pt = food.pivot_table(index=['City', 'Gender'], columns=['Item', 'Frequency'], values='Spends', aggfunc='mean')
```
